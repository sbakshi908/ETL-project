```python
import requests
from config import rapid_api_key
import pandas as pd
import pymongo
import datetime
from pymongo import MongoClient
```

```python
#ask user to enter the date for which they want to see the matches
user_date = input('Please enter a date in format "YYYY-MM-DD": ')
```

```python
#api request from the website and use the user date 

url = f'https://tennis-live-data.p.rapidapi.com/matches-by-date/{user_date}'
print(url)

headers = {
    'x-rapidapi-key': rapid_api_key,
    'x-rapidapi-host': "tennis-live-data.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers)

print(response.text)
```

```python
#get the response in json format
response_json = response.json()
response_json
```

```python
results = response_json['results'][0]['matches'][2]['result']['winner_id']
results
```

```python
#create empty list that will later be appended with dictionaries from the results
tennis_list = []

results = response_json['results']
for result in results:
    tourn_dict = {}
    
    
    tourn_dict['City'] = result['tournament']['city']
    tourn_dict['Name'] = result ['tournament']['name']
    tourn_dict['Surface'] = result ['tournament']['surface']
    tourn_dict['End_date'] = result ['tournament']['end_date']   
    tourn_dict['Start_date'] = result ['tournament']['start_date']
    tourn_dict['End_date'] = result ['tournament']['end_date']   
    tourn_dict['Matches'] = result ['matches'][0]['title']
    tourn_dict['Home_ID'] =  result ['matches'][0]['home_id']
    tourn_dict['Home_Player'] = result ['matches'][0]['home_player']
    tourn_dict['Away_ID'] =  result ['matches'][0]['away_id']
    tourn_dict['Away_Player'] = result ['matches'][0]['away_player']
    tourn_dict['Winner_ID'] = result['matches'][0]['result']
    
    
    print(tourn_dict)
    tennis_list.append(tourn_dict)
    

```

```python
tennis_list
```

```python
tourn_df = pd.DataFrame(tennis_list)
tourn_df
```
